```python
df=pd.DataFrame()
```

```python
for i in range(1,500):
    response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=69b759b45dd976dacb0d636e125ab312&language=en-US&page={}'.format(i))
    temp_df=pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df=df.append(temp_df,ignore_index=True)
```

```python
df.head()
```

```python
top_voting=df[df.sort_values("vote_count",ascending=False)["vote_count"]>=1000]
top_voting
```

```python
top_voting_movies=top_voting.sort_values("popularity",ascending=False).head(100)
top_voting_movies
```

```python
top_voting_movies["Movie_years"]=pd.to_datetime(top_voting_movies['release_date']).dt.year
top_voting_movies
```

```python
plt.plot(top_voting_movies.groupby("Movie_years").count().sort_index(ascending=True).reset_index()["Movie_years"]);
```

```python
#top_voting_movies[top_voting_movies["Movie_years"]==2022].sort_values("popularity",ascending=False).loc[:,"title"].iloc[0]

famous_movie=[]
for i in top_voting_movies["Movie_years"].value_counts().index:
    famous_movie.append(top_voting_movies[top_voting_movies["Movie_years"]==i].sort_values("popularity",ascending=False).loc[:,["title","Movie_years","popularity"]].iloc[0])
    
```

```python
famous_movie[0]
title=[]
year=[]
popu=[]
for i in range(len(famous_movie)):
    title.append(famous_movie[i][0])
    year.append(famous_movie[i][1])
    popu.append(famous_movie[i][2])
```

```python
x1=pd.DataFrame(title,columns=["Movie_Names"])
x2=pd.DataFrame(year,columns=["Movie_Year"])
x3=pd.DataFrame(popu,columns=["Movies_popularity"])
final_movie_data=pd.concat([x1,x2,x3],axis=1)
final_movie_data.sort_values("Movie_Year",ascending=False)
```

# Weather Page Web-Scrapping

```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date
```

```python
url="https://forecast.weather.gov/MapClick.php?lat=34.053570000000036&lon=-118.24544999999995#.Y49QoxTP2Cg"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html.parser')
week=soup.find(id='seven-day-forecast-container')
week
```

```python
items=week.find_all(class_='forecast-tombstone')
items[0].find(class_='period-name').text
days=[item.find(class_='period-name').text for item in items]
days
```

```python
items[0].find(class_="short-desc").text
description=[item.find(class_="short-desc").text for item in items]
description
```

```python
items[0].find(class_="temp temp-high").text
temperature=[item.find(class_="temp").text for item in items]
temperature
```

```python
weather=pd.DataFrame({"Days":days,
             "Temperature":temperature,
             'Condition':description})
weather
```

```python
weather['Temperature'][0].split(":")
```

```python
weather["Temperature_condition"]=[i.split(":")[0] for i in weather["Temperature"]]
weather
```

```python
weather["Fahrenheit"]=[int(i.split()[1]) for i in weather["Temperature"]]
```

```python
weather.drop(columns="Temperature",inplace=True)
```

```python
weather
```

```python
weather.columns
```

```python
plt.figure(figsize=(15,7))
sns.barplot("Days","Fahrenheit",data=weather,hue="Temperature_condition");
```
